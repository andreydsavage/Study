```python
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn import metrics
from sklearn.preprocessing import StandardScaler

from lifelines.utils import concordance_index
```

```python

PATH = '../data/'
data = pd.read_csv('data_from_0010.csv').set_index('user_id')
train = pd.read_csv(PATH+'train.csv').set_index("user_id")

data.fillna(0, inplace=True)
data.drop('count_trans_date_max', axis=1,inplace=True)



data = data.join(train, on="user_id", how="left")
                 
data
```

```python
data_train = data[data['target'].notna()]
data_test = data[data['target'].isna()]
```

```python

from sklearn.linear_model import LogisticRegression
X, y = data_train.iloc[:, :-2], data_train['target'] 
X = pd.DataFrame(data = StandardScaler().fit_transform(X),columns=X.columns)
clf = LogisticRegression(random_state=0,max_iter=100).fit(X, y)

log_reg_predictions = clf.predict_proba(X)[:,1]

fpr, tpr, thresholds = metrics.roc_curve(y, log_reg_predictions)
print('Log reg AUC = ',metrics.auc(fpr, tpr))


df_log_reg_importance = pd.DataFrame({'name': clf.feature_names_in_, 'imp':clf.coef_.flatten()}).sort_values(by='imp', ascending=False)
```

```python
from lifelines.fitters.weibull_aft_fitter import WeibullAFTFitter

wei = WeibullAFTFitter()

wei.fit(data_train, duration_col='time',  event_col='target')
survival_function = wei.predict_survival_function(data_train).T.mean(axis=1)

concordance_index(data_train['time'], survival_function, data_train['target'])
```

```python
from lifelines.fitters.log_logistic_aft_fitter import LogLogisticAFTFitter

logAFT = LogLogisticAFTFitter()

logAFT.fit(data_train, duration_col='time',  event_col='target')
survival_function = logAFT.predict_survival_function(data_train).T.mean(axis=1)

concordance_index(data_train['time'], survival_function, data_train['target'])
```

```python
from lifelines.fitters.log_normal_aft_fitter import LogNormalAFTFitter

normalAFT = LogLogisticAFTFitter()

normalAFT.fit(data_train, duration_col='time',  event_col='target')
survival_function = normalAFT.predict_survival_function(data_train).T.mean(axis=1)

concordance_index(data_train['time'], survival_function, data_train['target'])
```
